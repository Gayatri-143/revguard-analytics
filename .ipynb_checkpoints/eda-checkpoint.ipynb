```python
#total revenue loss
total_loss = df['Revenue_Loss'].sum()
print("Total Revenue Loss: " , total_loss)
```

```python
#Expected Revenue Loss
from sklearn.linear_model import LogisticRegression
y = df['Churn Value']
X = df.drop(columns=['Churn Value'])
X = pd.get_dummies(X, drop_first=True)
model = LogisticRegression(max_iter=1000)
model.fit(X, y)


df['Churn_Prob'] = model.predict_proba(X)[:,1]

df['Expected_Loss'] = df['Monthly Charges'] * df['Churn_Prob']
```

### Distribution

```python
sns.countplot(x='Churn Label', data=df)
plt.title("Churn Distribution")
plt.show()

print(df['Churn Label'].value_counts(normalize=True))
```

```python
df['Revenue_Loss'] = df['Monthly Charges'] * df['Churn Value']

print("Total Revenue Loss:", df['Revenue_Loss'].sum())
```

```python
num_cols = ['Tenure Months', 'Monthly Charges', 'Total Charges']

df[num_cols].hist(figsize=(12, 4))
plt.suptitle("Numerical Feature Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Churn Label', y='Tenure Months', data=df)
plt.title("Churn vs Tenure")
plt.show()
```

```python
# Low tenure customers churn more.
'''
 The boxplot indicates that customers who churn have significantly lower tenure compared to thosewho stay. 
 This suggests that customer retention is weakest in the early lifecycle stage, highlighting the importance of onboarding and early engagement strategies.
'''
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Churn Label', y='Monthly Charges', data=df)
plt.title("Churn vs Monthly Charges")
plt.show()
```

```python
# Higher charges -> higher churn tendency.
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Contract', hue='Churn Label', data=df)
plt.title("Churn vs Contract Type")
plt.show()
```

```python
# Month-to-month customers churn the most.
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='Payment Method', hue='Churn Label', data=df)
plt.xticks(rotation=45)
plt.title("Churn vs Payment Method")
plt.show()
```

```python
# Electronic chech have more churn rate 
'''
The analysis shows that customers paying via electronic check exhibit significantly higher churn 
compared to those using automatic payment methods such as bank transfer or credit card.
This suggests that customers with manual payment methods are more likely to leave, 
possibly due to lower convenience or engagement.
'''
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Internet Service', hue='Churn Label', data=df)
plt.title("Churn vs Internet Service")
plt.show()
```

```python
''' 
Fiber optic users show the highest churn rates, while DSL users have relatively lower churn, 
and customers without internet service exhibit minimal churn. 
This suggests that churn is strongly associated with the type of internet service subscribed.
'''
```

# Revenue-Based Insights

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Contract', y='Revenue_Loss', data=df)
plt.title("Revenue Loss by Contract Type")
plt.show()
```

```python
# 2 years contract have less revenue loss , while the month-to-month contract have high revenue loss rate
# Encouraging customers to switch from month-to-month to long-term contracts could significantly reduce churn and revenue loss.
```

### Key Insights:

```python
# Customers with month-to-month contracts have highest churn.
# Customers with low tenure are more likely to leave early.
# Higher monthly charges increase churn probability.
# Significant revenue loss comes from high-paying churn customers.
```
